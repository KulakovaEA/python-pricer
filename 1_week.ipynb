```python
from datetime import date, timedelta
import random
import numpy as np
import matplotlib.pyplot as plt

# Создание функции, которая будет считать кол-во дней 

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

# Объявление начала и конца исчесления дней
        
start_date_1 = date(2017, 1, 1)
end_date_1 = date(2017, 6, 1)
```

```python
# Генерация массива [[дата_1; своп пойнт_1], [дата_2; своп пойнт_2],...]

list_dates_and_swops = []
for single_date in daterange(start_date_1, end_date_1):
    list_dates_and_swops.append([single_date.strftime("%Y-%m-%d"),random.randint(1233, 15000)])
    
print(list_dates_and_swops)
```
